```python
def print_board(board):
  for row in board:
    print(" | ".join(row))
    print("-" * 5)

def check_winner(board,player):
  for row in board:
    if all(cell == player for cell in row):
      return True
  for col in range(3):
    if all(board[row][col] == player for row in range(3)):
      return True
  if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
    return True
  return False

def is_board_full(board):
  return all(all(cell != ' ' for cell in row) for row in board)

def play_game():
  board = [[' ' for _ in range(3)] for _ in range(3)]
  player = 'x'
  while True:
    print_board(board)
    row = int(input("Enter row (0, 1, 2): "))
    column = int(input("Enter column (0, 1, 2): "))
    if board[row][col] == ' ':
        board[row][col] = player
        if check_winner(board, player):
          print_board(board)
          print(f"player {player} wins!")
          break
        if is_board_full(board):
          print_board(board)
          print("It's a tie!")
          break
        player = 'o' if player == 'x' else 'x'
    else:
      print("That position is always taken. Try again.")

play_game()
```

```python
bank_data = {}

def create_account():
    name = input("Enter account holder's name: ")
    balance = float(input("Enter opening balance: "))
    account_number = len(bank_data) + 1
    bank_data[account_number] = {"name": name, "balance": balance}
    print("Account created successfully! Account number:", account_number)


def deposit():
    account_number = int(input("Enter account number: "))
    amount = float(input("Enter amount to deposit: "))
    if account_number in bank_data:
        bank_data[account_number]["balance"] += amount
        print("Deposit successful. Current balance:", bank_data[account_number]["balance"])
    else:
        print("Account not found.")

def withdraw():
    account_number = int(input("Enter account number: "))
    amount = float(input("Enter amount to withdraw: "))
    if account_number in bank_data:
        if bank_data[account_number]["balance"] >= amount:
            bank_data[account_number]["balance"] -= amount
            print("Withdrawal successful. Current balance:", bank_data[account_number]["balance"])
        else:
            print("Insufficient balance.")
    else:
        print("Account not found.")

def display_account():
    account_number = int(input("Enter account number: "))
    if account_number in bank_data:
        print("Account Holder's Name:", bank_data[account_number]["name"])
        print("Account Balance:", bank_data[account_number]["balance"])
    else:
        print("Account not found.")

def list_accounts():
    print("List of all accounts:")
    for account_number, account_info in bank_data.items():
        print("Account Number:", account_number)
        print("Account Holder's Name:", account_info["name"])
        print("Account Balance:", account_info["balance"])
        print()

def sort_accounts():
    sorted_accounts = sorted(bank_data.items(), key=lambda x: x[1]["balance"], reverse=True)
    print("Accounts sorted by balance:")
    for account_number, account_info in sorted_accounts:
        print("Account Number:", account_number)
        print("Account Holder's Name:", account_info["name"])
        print("Account Balance:", account_info["balance"])
        print()

def main():
    while True:
        print("\nBank Management System")
        print("1. Create Account")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Display Account Details")
        print("5. List All Accounts")
        print("6. Sort Accounts by Balance")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            create_account()
        elif choice == "2":
            deposit()
        elif choice == "3":
            withdraw()
        elif choice == "4":
            display_account()
        elif choice == "5":
            list_accounts()
        elif choice == "6":
            sort_accounts()
        elif choice == "7":
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
from abc import ABC, abstractmethod

class Car(ABC):
    def show(self):
        print("every cas has four wheels")

    @abstractmethod
    def speed(self):
        pass

class Maruti(Car):
    def speed(self):
        print("maximum speed 100km/hr")

class Suzuki(Car):
    def speed(self):
        print("maximum speed 110km/hr")

car1 = Maruti()
car1.show()
car1.speed()
car2 = Suzuki()
car2.show()
car2.speed()
```
